```python
# cleaning NAs
from pyspark.sql.types import DoubleType
from pyspark.sql.types import DoubleType

df = df.withColumn("ArrDelayDouble", df["ArrDelay"].cast(DoubleType()))
```

```python
display(df.groupBy('Month').avg('arrdelaydouble')) #NAs in the data
```

```python
## All are doing the same

from pyspark.sql.functions import *
display(df.select(['Year', 'ArrDelayDouble']).groupBy('Year').avg('ArrDelayDouble'))

display(df['Year', 'ArrDelayDouble']).groupBy('Year').avg('ArrDelayDouble')

display(df.groupBy('Month').avg('ArrDelayDouble'))
```

```python
# Filter

display(df.filter((df.Origin ==n 'SAN') & (df.Dest =='SFO'))
        .groupBy('DayOfWeek')
        .avg('arrdelaydouble'))
```

```python
from pyspark.sql.functions import concat, to_date

df = df.withColumn('DayDate', to_date(concat('Year', 'Month', 'DayofMonth'), 'yyyyMMdd'))

display(df.select(date_format('DayDate', 'E').alias('WeekDay'), 'arrdelaydouble', 'origin', 'DayOfWeek').filter(df.Origin == 'SAN').groupBy('WeekDay', 'DayOfWeek').agg(round(mean('arrdelaydouble'), 1).alias('AvgArrDelay')).sort('DayOfWeek'))
```

```python
airport_list = [row.Origin for row in df.select('Origin').distinct().limit(5).collect()]
print(airport_list)
```

```python
airportList = []

for row in df.select('Origin').distinct().limit(5).collect():
    airportList.append(row.Origin)
```

```python
#Rename

from pyspark.sql.functions import when

df = df.withColumn('State', when(col('Origin') == 'SAN', 'California')
                   .when(df.Origin =='LAX', 'California')
                   .when(df.Origin == 'SAN', 'California')
                   .when((df.Origin =='JFK') | (df.Origin == 'LGA') | (df.Origin == 'BUF'), "New York").otherwise('Other')
                        )
df.select('State').show()
```

```python
#Updating

df = df.withColumn('State', when(col('Origin') =='DFW', 'Texas'). when(col('Origin') == 'IAH', 'Texas').when(col('Origin') == 'DAL', 'Texas').otherwise(col('State')))
```

```python
display(df.groupBy('State').count())
```

```python
# Multiple DataFrames

df_dfw = df.select('DayOfMonth', 'Origin', 'Dest').filter(col('State') == 'California')

display(df_dfw.groupBy('Origin', 'Dest').count())
```

Joins

```python
df_airlines = spark.createDataFrame([
    ['AA', 'American Airlines'], 
    ['DL', 'Delta Airlines'],
    ['UA', 'United Airlines'],
    ['WN', 'Southwest Airlines']], ['ShortName', 'Fullname'])

df_hg= spark.createDataFrame([
    ['AA', 'Forth Worth', 'Texas'], 
    ['DL', 'Chicago', 'Illinois'],
    ['UA', 'Atlanta', 'Georgia'],
    ['FW', 'Swords', 'Ireland']], ['ShortName', 'City', 'State'])

df_cities= spark.createDataFrame([
    ['San Francisco'], 
    ['Miami'],
    ['Minneapolis']], ['City'])
```

```python
# JOIN

df_result = df_airlines.join(df_hg, df_airlines.ShortName == df_hg.ShortName)

display(df_result) # two times short name

#unng duplicated

df_result2 = df_airlines.join(df_hg.withColumnRenamed('ShortName', 'SN'), col("ShortName") == col('SN')) # the columns are named in the same way, then we don't have duplicated 
display(df_result2)

df_result3 = df_airlines.join(df_hg, ['ShortName'])

display(df_result3)
```

```python
# left, right, outer, cross_join

df_result_left = df_airlines.join(df_hg, ['ShortName'], 'left')

display(df_result_left)

df_result_right = df_airlines.join(df_hg, ['ShortName'], 'right')

display(df_result_right)

df_result_outer = df_airlines.join(df_hg, ['ShortName'], 'outer')

display(df_result_outer)
```

## Cross join

```python
df_result_cross = df_cities.crossJoin(df_cities.withColumnRenamed('City', 'Dest'))

display(df_result_cross)

# avoid duplicated rows

display(df_result_cross.where(col('City') != col('Dest')))
```
